```python
import pandas as pd
from scipy.stats import zscore

soju = pd.read_excel('Soju_rankings.xlsx')
```

```python
soju.columns = ['Soju Flavour','Amy','Rebekah','Christine','Vivian']
people = ['Amy','Rebekah','Christine','Vivian']
```

```python
standardized_score = pd.DataFrame()

# zscore is found per column i.e. per person rather than calculating group mean 

standardized_score['Soju Flavour'] = soju['Soju Flavour']
standardized_score[people] = zscore(soju[people])

standardized_score
```

```python
# apply zscores to the range 1-10 to make scores readable 

def stretch_range(person):
    smallest = min(person)
    biggest = max(person)
    a = 9/(-smallest + biggest)
    b = 1 - (a*smallest)

    person *= a
    person += b

    return person
```

```python
# apply function

for person in people:
    standardized_score[person] = stretch_range(standardized_score[person])
```

```python
# final scores

standardized_score = standardized_score.round(1)
standardized_score
```

```python
standardized_score.to_excel('soju_rankings_standardized.xlsx')
```
